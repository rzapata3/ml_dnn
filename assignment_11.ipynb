```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

cars = pd.read_csv('../data/Auto.csv', na_values = '?')
# inspect the data.

cars = cars.dropna()
# CHECK HOW MANY DIM ARE IN FOR INPUT
print(cars.describe())

# standardize the data.
scaler = StandardScaler()


# SET X VARIABLE
X_car = np.array(cars.loc[:,['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration','year','origin']])
standardized_x = scaler.fit_transform(X_car)

# SET Y VARIABLE
y_car = np.array(cars.mpg).reshape(-1,1)
standardized_y = scaler.fit_transform(y_car)


```

```python
#set up model for neural network
model_c = Sequential()
cars_loss_score = {}

# set up 5 Dense layers
# input layer.
model_c.add(Dense(1, input_dim=7, activation='sigmoid'))  #sigmoid, relu

# hidden layer
model_c.add(Dense(2, activation = 'relu'))
model_c.add(Dense(3, activation = 'tanh'))
model_c.add(Dense(3, activation = 'tanh'))    


# output layer
model_c.add(Dense(1, activation = 'sigmoid'))
sgd = SGD(lr=0.1)
model_c.compile(loss='binary_crossentropy', optimizer='sgd')
model_c.fit(standardized_x, standardized_y, batch_size=2, epochs=400)
cars_loss_score['cars'] = model_c.evaluate(standardized_x,standardized_y)
```
